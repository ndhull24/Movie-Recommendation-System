```python
#Creating the recommendation system
```

Create a matrix that has the user ids on one access and the movie title on another axis. Each cell will then consist of the rating the user gave to that movie. Note there will be a lot of NaN values, because most people have not seen most of the movies.

```python
df_new = df.pivot_table(index = 'user_id', columns='title', values ='rating')
df_new.head()
```

```python
#Finding the most rated movies with their average ratings

ratings.sort_values(('number_of_ratings'), ascending = False).head()
```

Lets create a correlation matrix of every movie with every other movie on user ratings. We will use the same matrix to find the best matches that relate to a particular movie and the result obtained will be added to the ratings dataframe of evry movie. Movies of which the matches could not be obtained using correlation, their value will be converted to "-".

```python
for i in ratings.index:
    # Get user ratings for the current movie
    movieUserRatings = df_new[i]
    
    # Compute correlations with all other movies
    similarToThatMovie = df_new.corrwith(movieUserRatings)
    
    # Create a DataFrame from the correlation results
    corr_toMovie = pd.DataFrame(similarToThatMovie, columns=['Correlation'])
    
    # Drop NaN values
    corr_toMovie.dropna(inplace=True)
    
    # Join with the number of ratings for each movie
    corr_toMovie = corr_toMovie.join(ratings['number_of_ratings'], how='inner')
    
    # Filter movies with more than 100 ratings and sort by correlation
    result = corr_toMovie[corr_toMovie['number_of_ratings'] > 100]\
        .sort_values('Correlation', ascending=False).head()

    # Check if there are at least 5 movies in the result
    if result['number_of_ratings'].count() >= 5:
        print(i)
        # Assign top recommendations (skip the first one since it's the same movie)
        try:
            ratings.loc[i, "FirstMovieRecommendation"] = result.iloc[1].name
            ratings.loc[i, "SecondMovieRecommendation"] = result.iloc[2].name
            ratings.loc[i, "ThirdMovieRecommendation"] = result.iloc[3].name
            ratings.loc[i, "FourthMovieRecommendation"] = result.iloc[4].name
        except IndexError as e:
            print(f"Not enough recommendations for movie {i}: {e}")
```

```python
#Check the results
ratings.head()
```

```python
ratings = ratings.fillna('-')
```

```python
#Save the ratings data for later use
ratings.to_csv('MovieRecommendations.csv', encoding = 'utf-8')
```
